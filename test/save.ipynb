# 連結してからプロットしたら変な線が入っている

```python
polo = poloniex.Poloniex()
chart_data1 = polo.returnChartData('USDT_BTC', period=300, start=time.time()-polo.DAY*99, end=time.time())
chart_data2 = polo.returnChartData('USDT_BTC', period=300, start=time.time()-polo.DAY*199, end=time.time()-polo.DAY*100)  #データ取得

df1 = pd.DataFrame(chart_data1)
df2 = pd.DataFrame(chart_data2)
df = pd.concat([df1,df2])

data = df.astype("float32")
data.plot(x='date', y='close')
```

# concatの引数のリストの順を変えてやればいい

```python
polo = poloniex.Poloniex()
chart_data1 = polo.returnChartData('USDT_BTC', period=300, start=time.time()-polo.DAY*99, end=time.time())
chart_data2 = polo.returnChartData('USDT_BTC', period=300, start=time.time()-polo.DAY*199, end=time.time()-polo.DAY*100)  #データ取得

df1 = pd.DataFrame(chart_data1)
df2 = pd.DataFrame(chart_data2)
df = pd.concat([df2,df1])

data = df.astype("float32")
data.plot(x='date', y='close')
```

# 4年強分のデータを取ってきてやる！！！！！！！！！！！！！！！！！

```python
polo = poloniex.Poloniex()
df_list=[]
n = 13

for j in range(n):
    i = n-j-1
    chart_data = polo.returnChartData('USDT_BTC', period=300, start=time.time()-polo.DAY*(i+1)*100, end=time.time()-polo.DAY*i*100)
    df_list.append(pd.DataFrame(chart_data))

df = pd.concat(df_list)

data = df.astype("float32")
data.plot(x='date', y='close')

data.to_pickle('4year_data.pkl')
```
